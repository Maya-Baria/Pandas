# Group by and Aggregating in Pandas

```python
Ice_creams = pd.read_csv(r"F:\Maya_Tech\Data_Analyst\Python\Flavors.csv")
Ice_creams
#print(Ice_creams.dtypes)
```

```python
group_by_frame = Ice_creams.groupby('Base Flavor').mean(numeric_only = True)
group_by_frame
```

```python
Ice_creams.groupby('Base Flavor').count()
```

```python
Ice_creams.groupby('Base Flavor').min()
```

```python
Ice_creams.groupby('Base Flavor').max()
```

```python
Ice_creams.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum']})
```

```python
Ice_creams.groupby(['Base Flavor', 'Liked']).mean(numeric_only = True)
```

```python
Ice_creams.groupby('Base Flavor').describe()
```
