# Creating Visualizations using Pandas Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"F:\Maya_Tech\Data_Analyst\Python\Sci-Fi_movies.csv")
df.head()
#print(df.dtypes)
```

```python
print(df.columns)
```

```python
# Visualize distribution of the ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of Ratings')
plt.xlabel('num_raters')
plt.ylabel('movie_rated')
plt.show()
```

```python
# Visualize relationship between rating and number of raters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='num_raters', y='rating', data=df, color='green')
plt.title('Rating vs Number of Raters')
plt.xlabel('num_raters')
plt.ylabel('rating')
plt.show()
```

```python
Rate_count = df["movie_rated"].value_counts()
Rate_count
```

```python
Count_year = df["year"].value_counts()
Count_year
```

```python
#Line graph
plt.figure(figsize = (10,5))
sns.lineplot(x=df['year'].value_counts().index,y=df['year'].value_counts())
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation=45, ha="right")
sns.countplot(x='movie_rated',data=df)
plt.show()
```
